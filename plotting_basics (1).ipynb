## Line Plots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  
```

```python
data = pd.read_csv('Teams.csv')
```

```python
plt.plot(data['yearID'],data['CG'])
```

This plot shows the relationship between the number of complete games in a season for a team, and the year. We can see that as time went on, the number of complete games in a year fell.

## Scatter

Let's look at one other common kind of plot first, the scatter. This works much the same as `.plot()` and is called with `.scatter()` instead.

```python
plt.scatter(x=data['HR'], y=data['SO'])
plt.show()
```

This plot compares the number of home runs a team hits a year with the number of times the team strikes out. As a team hits more home runs, the amount of strike outs they have also increases.

## Histogram

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Plot them as a histogram.
plt.hist(x=data['ERA']) 
plt.show()
```

This histogram shows the distribution of ERA (Earned Runs Average) of teams in the dataset. This plot shows us that the data is concentrated around 3 - 4 ERA in a year.

## Boxplot

```python
# Set the random seed to keep the example consistent.
np.random.seed(111)

# Sample data.
x = data['W']

# Generate and display the box plot.
plt.boxplot(x)
plt.show()
```

This plot makes a boxplot of the total number of wins in a year for a team. The boxplot shows the median number of wins in a season is around 80. The large number of outliers are probably a result of seasons with less games played than modern seasons being included.
