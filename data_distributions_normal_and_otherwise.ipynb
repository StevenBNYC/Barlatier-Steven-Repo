```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exponential = np.random.exponential(1, 100)
plt.axvline(exponential.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(exponential.mean() + exponential.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(exponential.mean()-exponential.std(), color='b', linestyle='dashed', linewidth=2) 

plt.hist(exponential)
```

```python
negative_binomial = np.random.negative_binomial(1,.5, 100)
plt.axvline(negative_binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(negative_binomial.mean() + negative_binomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(negative_binomial.mean()-negative_binomial.std(), color='b', linestyle='dashed', linewidth=2) 

plt.hist(negative_binomial)
```

```python
logistic = np.random.logistic(10,1, 100)
plt.axvline(logistic.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(logistic.mean() + logistic.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(logistic.mean()-logistic.std(), color='b', linestyle='dashed', linewidth=2) 

plt.hist(logistic)
```

```python
triangular = np.random.triangular(-5,0,5, 100)
plt.axvline(triangular.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(triangular.mean() + triangular.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(triangular.mean()-triangular.std(), color='b', linestyle='dashed', linewidth=2) 

plt.hist(triangular)
```

```python
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='b')
plt.axvline(x=mean+sd, color='b')
plt.axvline(x=mean-sd, color='b')
plt.show()
```
